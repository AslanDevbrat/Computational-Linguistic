## Reading text file

```python
# Importing the files

import pandas as pd
import numpy as np
import sys
```

```python
# Reading the text corpus
file1 = open("mother-goose-corpus.txt","r") 
```

```python
# array to hold every line as 1 element from the corpus
arr=[]

with open("mother-goose-corpus.txt", "r") as a_file:
      for line in a_file:
        stripped_line = line.strip()
        arr.append(stripped_line)
        print(stripped_line)
```

## Q1 Creating vocabulary set

```python
# dictionary to hold {token:frequency} for vocabulary of the corpus 
vocabluray_dict = {}

# Looping through the corpus text an counting the occurance of 
# each unique word
## NOTE::
## I have taken <s> and </s> as two unique words
for sentence in arr:
    for word in sentence.split():
        if word in vocabluray_dict:
            vocabluray_dict[word]+=1
        else:
            vocabluray_dict[word] =1

print('Vocabulary :=', [*vocabluray_dict])        
```

## Unigram Prob            
    

```python
# dictionary to hold unigram and their Probability as  key- value pair
unigram_dict = dict()

import operator

for item in vocabluray_dict:
    unigram_dict[item]= vocabluray_dict[item]/sum(vocabluray_dict.values())

#sorting the unigram dict in reverse order to get the most frequent unigram
sorted_x = sorted(unigram_dict.items(), key=operator.itemgetter(1),reverse=True)

print("unigram with highest probability is  \'",sorted_x[0][0],'\'with probability ',sorted_x[0][1])
print("unigram with second highest Probability \'",sorted_x[1][0], '\'with probability ',sorted_x[1][1])
```

```python
vocabluray_dict.pop('</s>')
vocabluray_dict.pop('<s>')

```

```python
#print(vocabluray_dict)
frequencies = pd.DataFrame(vocabluray_dict,index=['Frequency'])
frequencies.head()
```

```python
# using zip() + split() + list comprehension 
# for Bigram formation 
bigrams = set(i for j in arr  
       for i in zip(j.split(" ")[:-1], j.split(" ")[1:]))
#print(res)
```

```python
# Dataframe to hold bigram, bigram count and their Probability
df_bigram = pd.DataFrame(None, columns = ['Bigram', 'Bigram Count','Bigram Prob'],index=range(len(bigrams)))


for i,x in enumerate(bigrams) :
    #print(x[0],x[1])
    count=0
    for sentence in arr:
        #print(sentence)
        
        count+=sentence.count(x[0]+" "+x[1])
    
    # setting value to the respective column in the Dataframe
    df_bigram.at[i,'Bigram'] = x[0]+" "+x[1]
    df_bigram.at[i,'Bigram Prob'] = count/vocabluray_dict[x[0]]
    df_bigram.at[i,'Bigram Count'] = count
    #print('count of' ,x[0]+" "+x[1],' is',count)
    
df_bigram
```

```python
# Dataframe to hold MLE for the bigram
df_mle = pd.DataFrame (0.0, columns = [*vocabluray_dict],index=[*vocabluray_dict])
#df_mle.head(13)
pd.options.display.precision = 2
bigram_list = df_bigram.Bigram
for i,bigram in enumerate(bigram_list):
    
    if '<s>' in bigram or '</s>' in bigram:
        #print(bigram)
        continue
    words_of_bigram = bigram.split()
    #print(words_of_bigram)
    x = (df_bigram.at[i,'Bigram Count']/frequencies.at['Frequency',words_of_bigram[0]]).round(2)
    df_mle.at[words_of_bigram[0], words_of_bigram[1]] = float(x)

df_mle.head(19)
# This data fram show MLE for the seen data 
```

## Q5 Calculating Laplace Smoothing


```python
# Dataframe to hold Laplace Smoothed matrix
df_laplace = df_mle = pd.DataFrame (1.000, columns = [*vocabluray_dict],index=[*vocabluray_dict])
bigram_list = df_bigram.Bigram
for i,bigram in enumerate(bigram_list):
    
    if '<s>' in bigram or '</s>' in bigram:
        #print(bigram)
        continue
    words_of_bigram = bigram.split()
    #print(words_of_bigram)
    x = (df_bigram.at[i,'Bigram Count'])/(frequencies.at['Frequency',words_of_bigram[0]] +len([*vocabluray_dict]))
    #print(x)
    df_laplace.at[words_of_bigram[0], words_of_bigram[1]] = x

# dealing with value 1
#Calculation for the Laplace Smoothing for the unseen Data
unseen_bigram_dict = dict()
for index, row in df_laplace.iterrows():
    for column in [*vocabluray_dict]:
        if row[column] == 1:
            df_laplace.at[row.name,column] = 1/(len([*vocabluray_dict])+frequencies.at['Frequency',row.name])
            unseen_bigram_dict[row.name+" "+column] = df_laplace.at[row.name,column] 

```
