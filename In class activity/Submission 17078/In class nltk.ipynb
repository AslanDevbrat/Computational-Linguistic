```python
patterns = [(r'.*ing$', 'VBG'),       # gerunds
    (r'.*ed$', 'VBD'),                # simple past
    (r'.*es$', 'VBZ'),                # 3rd singular present
    (r'.*ould$', 'MD'),               # modals
    (r'.*\'s$', 'NN$'),               # possessive nouns
    (r'.*s$', 'NNS'),                 # plural nouns
    (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),  # cardinal numbers
    (r'.*', 'NN')                     # nouns (default)
           ]
patterns
```

```python
regexp_tagger = nltk.RegexpTagger(patterns)
regexp_tagger.tag(tokens)
```

```python
regexp_tagger.evaluate(brown_tagged_sents)*100
```

```python
 #Cutoff = Number of instances of training data the tagger must see in order not to use the backoff tagger
#backoff = Another tagger which this tagger will consult when it is unable to tag a word


bigram_tagger = nltk.BigramTagger(brown, cutoff=0, backoff = None)
```

```python
sent = 'I eat an apple and two mangoes .'.split()
# sent = 'Colorless green ideas sleep furiously .'.split()
bigram_tagger.tag(sent)
```

```python
#Accuracy

print(bigram_tagger.evaluate(brown_tagged_sents)*100)
```

```python
#Interpolated Tagger

t0 = nltk.DefaultTagger('NN')
t1 = nltk.UnigramTagger(brown, backoff=t0)
t2 = nltk.BigramTagger(brown, backoff=t1)
t2.tag(sent)
```

```python
#Accuracy

print(t2.evaluate(brown_tagged_sents)*100)
```
