## Q1

The reason we get the probability -inf  for the OOV words is that, The Srilm assign a probability of 0 to the out of vocabulary words and log(0) = - infinity, there fore the -inf we got is actually the log Probability. 

- an **unk** token is assumed to take the place of the OOV word and no back-off may be necessary if a corresponding N-gram containing **unk** is found in the LM.

- -addsmooth deal with OOV words by adding a constant to **unk** words

This can be fixed by using **-addsmooth 1**
